```python
def Game_Introduction():
    
    print ("Welcome to the Guess the hidden number game,\nHere in the breif introduction about the game:\n\nIts a very simple game,\nAll you have to do is guess the hidden number.\n\nIn the game after your every guess or input you will told if that guess was further or closer to the\nhidden number compared to your previous guess.\n\nIf your guess came out closer to the hidden number compared to your previous guess then the game\nwill tell you, 'Warmer' and 'Colder' if further\n\nOn your first turn if your guess came out within the '10' of that nummber, It will tell you 'Warm' and 'Cold' if it was further than 10")
    
    print ("\nAnd when you have guessed the number correctly, You'll be told your answer is correct and\nalong with how many guess it took.\n")

Game_Introduction()

from random import randint

result = randint(1,100)


mylist = [0]

Guess = int(input('Your Guess: '))

index_count = 0

if Guess == result - 10 or Guess == result + 10 or Guess == result:
    print('Warm')
elif Guess > 100 or Guess < 1:
    print('Out of Bounds')
else:
    print('Cold')
        
while Guess != result:
    Guess = int(input('\nYour Guess: '))
    if Guess > 100 or Guess < 1:
        print('Out of Bounds')
        
    elif Guess == result:
        print(f'\nGood Job, Your guess {Guess} was the hidden number and it took you {len(mylist)} guesses to find the right answer.')
        
    elif abs(Guess - result) < abs(mylist[index_count] - result):
        print('Warmer')

    else:
        print('Colder')
    
    mylist.append(Guess)
    index_count += 1
    
if Guess == result:
        print(f'\nGood Job, Your guess {Guess} was the hidden number and it took you {len(mylist)} guesses to find the right answer.')
```
